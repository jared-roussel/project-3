```python
import pandas as pd
```

```python
data_path_one = '../Resources/superbowl.csv'
data_path_two = '../Resources/Major_Sport_Venues.csv'
data_path_three = '../Resources/Fortune_500_Corporate_Headquarters.csv'
```

```python
data_one = pd.read_csv(data_path_one)
```

```python
date = data_one['Date']
sb = data_one['SB']
winner = data_one['Winner']
win_pts = data_one['Winner Pts']
loser = data_one['Loser']
los_pts = data_one['Loser Pts']
mvp = data_one['MVP']
stadium = data_one['Stadium']
city = data_one['City']
state = data_one['State']

superbowl_df = pd.DataFrame({'date': date,
                            'superbowl_num': sb,
                            'winner': winner,
                            'winner_points': win_pts,
                            'loser': loser,
                            'loser_points': los_pts,
                            'mvp': mvp,
                            'stadium': stadium,
                            'city': city,
                            'state': state})

superbowl_df['year'] = superbowl_df['date'].str[-4:]
superbowl_df['stadium'] = superbowl_df['stadium'].str.lower()
```

```python
data_two = pd.read_csv(data_path_two)
```

```python
stadium_coo = data_two['NAME']
stadium_lat = data_two['LATITUDE']
stadium_lng = data_two['LONGITUDE']

stadium_coords = pd.DataFrame({'stadium': stadium_coo,
                              'latitude': stadium_lat,
                              'longitude': stadium_lng})

stadium_coords['stadium'] = stadium_coords['stadium'].str.lower()
stadium_coords
```

```python
combined_superbowl = pd.merge(superbowl_df, stadium_coords, on = 'stadium', how = 'inner')
combined_superbowl.sort_values('year')
#cleaned_superbowl = combined_superbowl.drop_duplicates(subset = ['year'], keep = 'first')
combined_superbowl.reset_index().drop(columns = 'index')
combined_superbowl
```

```python
data_three = pd.read_csv(data_path_three)
data_three.head()
```

```python
lat_three = data_three['LATITUDE']
lng_three = data_three['LONGITUDE']
rank = data_three['RANK']
company = data_three['NAME']
address = data_three['ADDRESS']
city_three = data_three['CITY']
state_three = data_three['STATE']
employees = data_three['EMPLOYEES']
revenues = data_three['REVENUES']
profit = data_three['PROFIT']

company_df = pd.DataFrame({'latitude': lat_three,
                          'longitude': lng_three,
                          'rank': rank,
                          'company_name': company,
                          'address': address,
                          'city': city_three,
                          'state': state_three,
                          'employee_count': employees,
                          'revenue': revenues,
                          'profit': profit})

company_df = company_df.sort_values('rank').reset_index().drop(columns = 'index')
company_df['city'] = company_df['city'].str.title()
company_df
```

```python
company_df_short = pd.DataFrame({'city': city_three,
                                'state': state_three})
company_df_short['city'] = company_df_short['city'].str.title()
```

```python
combined_superbowl.to_csv('../Resources/cleaned_superbowl.csv')
company_df.to_csv('../Resources/cleaned_fortune500.csv')
superbowl_df.to_csv('../Resources/missing_superbowl_coordinates.csv')
```
